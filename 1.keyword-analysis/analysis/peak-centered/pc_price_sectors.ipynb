# 1. Peak-centered correlation on different sectors

```python
def ccf(x, y, lag_max = 100):
    result = ss.correlate(y - np.mean(y), x - np.mean(x), method='direct') / (np.std(y) * np.std(x) * len(y))
    length = (len(result) - 1) // 2
    lo = length - lag_max
    hi = length + (lag_max + 1)

    return result[lo:hi]
```

```python
corr_list = {sector:list() for sector in pd.unique(full_company_df['sector'].values)}
lag = 7
num_peaks = 20
scale_columns = ['volume', 'price', 'ratio']


for idx1, row1 in full_company_df.iterrows():
    if row1['type'] == 'kospi':
        filepath= KOSPI_DIR + row1['code'] + '.csv'
    else:
        filepath= KOSDAQ_DIR + row1['code'] + '.csv'
    
    try:
        df = pd.read_csv(filepath)
    except FileNotFoundError:
        continue
    df[scale_columns] -= df[scale_columns].min()
    df[scale_columns] /= df[scale_columns].max()
    df = df.dropna()
    
    peaks = ss.find_peaks(df['ratio'].values, distance=7)[0]
    top_peaks = df.iloc[peaks,:]
    top_peaks = top_peaks.nlargest(num_peaks, 'ratio')
    # print(top_peaks)
    
    
    top_timeframes = {}
    # print(top_ten)
    
    # Create timeframe for each peak
    for idx, row in top_peaks.iterrows():
        # print(row)
        start = idx-lag 
        end = idx+lag
        if start < 0 or end > len(df.index):
            continue
        # print("Start: ", start, "End: ", end)
        timeframe = df[start:end]
        top_timeframes[row['period']] = timeframe
        
    # print(top_timeframes)
        
    # Calculate cross correlation
    top_ncc = {}
    for peak, timeframe in top_timeframes.items():
        prices = timeframe['price'].values
        ratio = timeframe['ratio'].values    
        ncc = ccf(prices,ratio,lag_max=lag)
        max_idx = np.argmax(abs(ncc))
        max_ncc = ncc[max_idx]
        max_lag = max_idx - lag 
        top_ncc[peak] = {
            "ncc": list(ncc),
            "max": [max_ncc, max_lag]
        }
        corr_list[row1['sector']].append(max_ncc)
```

```python
# Convert Korean to English

kor = ['에너지', '소재', '산업재', '경기관련소비재', '필수소비재', '건강관리', '금융', 'IT', '커뮤니케이션서비스', '유틸리티']

eng = ['energy', 'materials', 'industrial goods', 'economy-related consumer goods', 'essential consumer goods',
       'healthcare', 'finance', 'IT', 'communication service', 'utility']

kor2eng = dict(zip(kor,eng))
```

```python
# Dataframe For Violin Plot

new_list = []
for sector in corr_list:
    for value in corr_list[sector]:
        new_list.append({'sector': kor2eng[sector], 'value': value})
        
new_df = pd.DataFrame(new_list)
```

```python
# Draw Violin plot For Sector Comparison

plt.figure(figsize=(22, 8))
ax = sns.violinplot(x="sector", y="value", data=new_df)
```
