```python
import os
import pandas as pd
import shutil
```

```python
KOSPI_PRICE_DIR = "./kospi/price/"
KOSPI_KEYWORD_DIR = "./kospi/keyword/"
KOSPI_COMBINED_DIR = "./kospi/combined/"

KOSDAQ_PRICE_DIR = "./kosdaq/price/"
KOSDAQ_KEYWORD_DIR = "./kosdaq/keyword/"
KOSDAQ_COMBINED_DIR = "./kosdaq/combined/"



PRICE_DIR = "./price/"
KEYWORD_DIR = "./keyword/"
COMBINED_DIR = "./combined/"


price_list = os.listdir(PRICE_DIR)
keyword_list = os.listdir(KEYWORD_DIR)
combined_list = os.listdir(COMBINED_DIR)
```

```python
# KOSPI
kospi = pd.read_csv('clean_code_comp.csv')
# KOSDAQ
kosdaq = pd.read_csv('clean_code_comp_kosdaq.csv')
```

```python
kospi['종목코드'] = kospi['종목코드'].map("{:06d}".format)
kosdaq['종목코드'] = kosdaq['종목코드'].map("{:06d}".format)
```

```python
kospi.head()
```

```python
kospi_dict = dict(zip(kospi['종목코드'].values, kospi['기업명'].values))
kosdaq_dict = dict(zip(kosdaq['종목코드'].values, kosdaq['기업명'].values))
```

```python
sorted(os.listdir(KOSDAQ_COMBINED_DIR))
```

```python
def move_files(file_list, src_dir, dest_dir, company_dict):
    for file in file_list:
        if file[:-4] in company_dict:
            src = src_dir + file
            dest = dest_dir + file
            shutil.move(src, dest)
```

```python
# KOSPI Files

# Move Price Files
# move_files(price_list, PRICE_DIR, KOSPI_PRICE_DIR,kospi_dict)

# Move Keyword Files
move_files(keyword_list, KEYWORD_DIR, KOSPI_KEYWORD_DIR, kospi_dict)

# Move Combined Files
move_files(combined_list, COMBINED_DIR, KOSPI_COMBINED_DIR, kospi_dict)
```

```python
# KOSDAQ Files

# Move Price Files
move_files(price_list, PRICE_DIR, KOSDAQ_PRICE_DIR, kosdaq_dict)

# Move Keyword Files
move_files(keyword_list, KEYWORD_DIR, KOSDAQ_KEYWORD_DIR, kosdaq_dict)

# Move Combined Files
move_files(combined_list, COMBINED_DIR, KOSDAQ_COMBINED_DIR, kosdaq_dict)
```
